```markdown
## 3. データクリーニングと前処理

「実務の8割はデータ加工」という言葉の通り、取得したデータはそのままでは分析に適さないことが多々あります。ここでは、特に`Price_Range_USD`カラムが文字列形式で格納されているため、これを数値として扱えるようにクレンジングし、分析に必要な新たな特徴量を作成します。

### 3.1 `Price_Range_USD`カラムの数値化

`Price_Range_USD`カラムは「$X-$Y」の形式で価格帯が記述されています。この文字列から最小価格と最大価格を抽出し、新しい数値型のカラムとして`Min_Price_USD`と`Max_Price_USD`を作成します。さらに、その平均値として`Average_Price_USD`も計算し、今後の分析で利用します。
```

````python
# Price_Range_USD カラムのクリーンアップと数値化
def parse_price_range(price_range_str):
    if isinstance(price_range_str, str):
        # '$'と','を削除し、'-'で分割
        cleaned_str = price_range_str.replace('$', '').replace(',', '')
        parts = cleaned_str.split('-')
        if len(parts) == 2:
            try:
                min_price = float(parts[0])
                max_price = float(parts[1])
                return min_price, max_price
            except ValueError:
                return None, None # 数値変換エラー
        elif len(parts) == 1: # 単一価格の場合
            try:
                price = float(parts[0])
                return price, price
            except ValueError:
                return None, None
    return None, None # 無効な形式

# 新しいカラム 'Min_Price_USD' と 'Max_Price_USD' を作成
df[['Min_Price_USD', 'Max_Price_USD']] = df['Price_Range_USD'].apply(lambda x: pd.Series(parse_price_range(x)))

# Average_Price_USD を計算 (欠損値がある場合は計算しない)
df['Average_Price_USD'] = df[['Min_Price_USD', 'Max_Price_USD']].mean(axis=1)

# クリーニング後のデータ確認
print("### 'Price_Range_USD' クリーニング後のデータフレームの先頭5行:\n")
print(df[['Price_Range_USD', 'Min_Price_USD', 'Max_Price_USD', 'Average_Price_USD']].head())
print("\n" + "="*50 + "\n")

print("### 新しい数値カラムの基本情報:\n")
df[['Min_Price_USD', 'Max_Price_USD', 'Average_Price_USD']].info()
print("\n" + "="*50 + "\n")

# 欠損値の確認（クリーンアップで発生する可能性もあるため）
print("### クリーニング後の欠損値の確認:\n")
print(df[['Min_Price_USD', 'Max_Price_USD', 'Average_Price_USD']].isnull().sum())

# もし価格範囲のパースに失敗した行があれば、その行を表示して原因を特定できるようにする。
# 例えば、単一価格の形式、あるいは予期せぬ形式がある場合。
failed_to_parse = df[df['Min_Price_USD'].isnull() & df['Price_Range_USD'].notnull()]
if not failed_to_parse.empty:
    print("\n### 価格範囲のパースに失敗した行:\n")
    print(failed_to_parse[['Product_Name', 'Price_Range_USD']])
    print("\nこれらの行は、`Price_Range_USD`カラムの予期せぬフォーマットが原因である可能性があります。今回は平均価格で補完せず、分析から除外するか、手動で修正することを検討します。")
````

```markdown
## 4. 利益関連指標の作成と課題の特定

データクリーニングが完了し、`Average_Price_USD`が利用可能になりました。次に、収益最大化の分析軸となる「利益」を算出します。このデータセットにはコスト情報がないため、ここでは仮に売上に対する粗利益率を設定し、推定利益を算出します。この推定利益を基に、カテゴリ別のパフォーマンスや価格帯との関係などを分析し、課題を特定します。

### 4.1 推定利益の計算

ここでは仮に粗利益率を **30%** と設定し、`Estimated_Profit_in_2025_USD`カラムを作成します。この仮定は分析の初期段階であり、実際のビジネスにおいては詳細なコストデータに基づいて調整されるべき点です。
```

````python
# 粗利益率を仮定 (例: 30%)
PROFIT_MARGIN = 0.30

# 推定利益を計算
df['Estimated_Profit_in_2025_USD'] = df['Estimated_Revenue_in_2025_USD'] * PROFIT_MARGIN

print(f"仮定粗利益率: {PROFIT_MARGIN*100}% で推定利益を計算しました。\n")
print("### 推定利益カラムを追加したデータフレームの先頭5行:\n")
print(df[['Estimated_Revenue_in_2025_USD', 'Estimated_Profit_in_2025_USD']].head())
print("\n" + "="*50 + "\n")

print("### 新しい数値カラムの基本情報:\n")
df[['Estimated_Profit_in_2025_USD']].info()
print("\n" + "="*50 + "\n")

print("### 推定利益の統計量:\n")
print(df['Estimated_Profit_in_2025_USD'].describe())
````

```markdown
### 4.2 カテゴリ別・サブカテゴリ別パフォーマンス分析

ここでは、`Category`と`Subcategory`ごとに、`Estimated_Revenue_in_2025_USD`、`Estimated_Profit_in_2025_USD`、`Estimated_Total_Units_Sold_in_2025`、そして`Average_Price_USD`の平均値を集計し、ビジネスにおける主要な課題を特定するための洞察を得ます。

**なぜこの分析を行うのか:**
ビジネスにおいて、どの製品カテゴリやサブカテゴリが収益の主要な源泉であるかを理解することは非常に重要です。収益性と販売数を多角的に評価することで、リソース配分の最適化、マーケティング戦略の調整、そして新製品開発の方向性を定めるための基礎情報となります。例えば、収益は高いが販売数が少ないカテゴリがあれば高価格帯戦略を、販売数が多いが利益率が低いカテゴリがあればコスト削減や価格見直しを検討するヒントになります。


````python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

sns.set_style("whitegrid")

# 日本語フォントの設定
# 利用可能な日本語フォントを探し、設定する
# Colab環境では'TakaoGothic'や'Noto Sans CJK JP'などが利用可能です
# 適切なフォントが見つからない場合は、フォントのインストールが必要になる場合があります。
# ここではGoogle Colabにデフォルトでインストールされている可能性のあるフォントを試します。

# まず、利用可能なフォントをリストアップ
# font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
# jp_font = [f.name for f in fm.fontManager.ttflist if 'IPAPGothic' in f.name or 'Noto Sans CJK JP' in f.name or 'DejaVu Sans' in f.name]

# Colabの環境に合わせて、よく使われる日本語フォントを設定
# 例えば、'Noto Sans CJK JP' や 'IPAexGothic' など
if 'Noto Sans CJK JP' in [f.name for f in fm.fontManager.ttflist]:
    plt.rcParams['font.family'] = 'Noto Sans CJK JP'
elif 'IPAexGothic' in [f.name for f in fm.fontManager.ttflist]:
    plt.rcParams['font.family'] = 'IPAexGothic'
else:
    print("Warning: Japanese font not found. Please install one (e.g., !apt-get install -y fonts-ipafont-gothic).")
    plt.rcParams['font.family'] = 'DejaVu Sans' # fallback to default

plt.rcParams['font.size'] = 12
plt.rcParams['axes.unicode_minus'] = False # マイナス記号を正しく表示

# カテゴリごとの集計
category_summary = df.groupby('Category').agg(
    Total_Revenue=('Estimated_Revenue_in_2025_USD', 'sum'),
    Total_Profit=('Estimated_Profit_in_2025_USD', 'sum'),
    Total_Units_Sold=('Estimated_Total_Units_Sold_in_2025', 'sum'),
    Average_Price=('Average_Price_USD', 'mean'),
    Product_Count=('Product_ID', 'count')
).sort_values(by='Total_Profit', ascending=False)

print("### カテゴリ別パフォーマンスサマリー (利益順):\n")
print(category_summary)
print("\n" + "="*80 + "\n")

# サブカテゴリごとの集計 (上位10件)
subcategory_summary = df.groupby('Subcategory').agg(
    Total_Revenue=('Estimated_Revenue_in_2025_USD', 'sum'),
    Total_Profit=('Estimated_Profit_in_2025_USD', 'sum'),
    Total_Units_Sold=('Estimated_Total_Units_Sold_in_2025', 'sum'),
    Average_Price=('Average_Price_USD', 'mean'),
    Product_Count=('Product_ID', 'count')
).sort_values(by='Total_Profit', ascending=False).head(10)

print("### サブカテゴリ別パフォーマンスサマリー (利益順、上位10件):\n")
print(subcategory_summary)
print("\n" + "="*80 + "\n")

# 視覚化: カテゴリ別総利益
plt.figure(figsize=(14, 7))
sns.barplot(x=category_summary.index, y='Total_Profit', data=category_summary, palette='viridis')
plt.title('カテゴリ別 総推定利益 (2025年)', fontsize=16)
plt.xlabel('カテゴリ', fontsize=14)
plt.ylabel('総推定利益 (USD)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 視覚化: 上位サブカテゴリ別総利益
plt.figure(figsize=(14, 7))
sns.barplot(x=subcategory_summary.index, y='Total_Profit', data=subcategory_summary, palette='magma')
plt.title('上位サブカテゴリ別 総推定利益 (2025年)', fontsize=16)
plt.xlabel('サブカテゴリ', fontsize=14)
plt.ylabel('総推定利益 (USD)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 視覚化: カテゴリ別 平均価格と総販売数
fig, ax1 = plt.subplots(figsize=(16, 8))

color = 'tab:red'
ax1.set_xlabel('カテゴリ', fontsize=14)
ax1.set_ylabel('平均価格 (USD)', color=color, fontsize=14)
sns.barplot(x=category_summary.index, y='Average_Price', data=category_summary, ax=ax1, color=color, alpha=0.6, label='平均価格')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(category_summary.index, rotation=45, ha='right')

ax2 = ax1.twinx() # 平均価格と販売数を一つのグラフに
color = 'tab:blue'
ax2.set_ylabel('総販売数', color=color, fontsize=14)
sns.lineplot(x=category_summary.index, y='Total_Units_Sold', data=category_summary, ax=ax2, color=color, marker='o', linewidth=2, label='総販売数')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('カテゴリ別 平均価格と総販売数 (2025年)', fontsize=16)
fig.tight_layout() # レイアウトの調整
plt.show()
````

## サブカテゴリごとの製品数と販売数の相関分析

「Beauty & Skincare」カテゴリのデータを抽出し、サブカテゴリごとに製品数と総販売数を集計します。これらの集計結果を用いて散布図を作成し、製品数と販売量の間にどのような相関があるかを視覚的に分析します。また、ピアソン相関係数を計算して数値的に相関の強さを評価します。


## Summary:

### Data Analysis Key Findings
The analysis focused on the 'Beauty & Skincare' category to understand the relationship between product count and total sales volume across its subcategories.

*   Data for 'Beauty & Skincare' was extracted and aggregated by subcategory to compute the number of products and total sales volume.
*   A scatter plot was generated to visually represent the correlation between product count and sales volume for each subcategory.
*   The Pearson correlation coefficient was calculated to quantify the strength and direction of the linear relationship between product count and total sales. (Specific numerical findings are not provided in the process description.)

### Insights or Next Steps
*   Examine the scatter plot and the calculated Pearson correlation coefficient for specific subcategories to identify strong positive or negative correlations, and outliers. This could indicate opportunities for product portfolio optimization or marketing focus.
*   Investigate subcategories with high product counts but disproportionately low sales, or vice versa, to understand underlying causes such as market demand, pricing strategies, or product quality.


**Reasoning**:
I will extract the 'Beauty & Skincare' data, aggregate it by subcategory to get product count and total sales, then visualize this relationship with a scatter plot and calculate the Pearson correlation coefficient, as per the instructions.



````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. 'Beauty & Skincare' カテゴリのデータを抽出
beauty_skincare_df = df[df['Category'] == 'Beauty & Skincare'].copy()

# 2. サブカテゴリごとに製品数と総販売数を集計
subcategory_sales_analysis = beauty_skincare_df.groupby('Subcategory').agg(
    Product_Count=('Product_ID', 'count'),
    Total_Sales=('Estimated_Revenue_in_2025_USD', 'sum')
).reset_index()

# 3. 集計結果を表示
print("### 'Beauty & Skincare' カテゴリのサブカテゴリ別製品数と総販売数:")
print(subcategory_sales_analysis)
print("\n" + "="*80 + "\n")

# 4. 散布図を作成
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='Product_Count',
    y='Total_Sales',
    data=subcategory_sales_analysis,
    hue='Subcategory', # サブカテゴリごとに色分け
    size='Total_Sales', # 売上高に応じて点のサイズを変更
    sizes=(50, 1000), # 点のサイズ範囲
    legend='full' # 凡例を表示
)
plt.title('Beauty & Skincare: サブカテゴリ別 製品数 vs. 総販売数 (2025年)', fontsize=16)
plt.xlabel('製品数', fontsize=14)
plt.ylabel('総販売数 (USD)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()

# 5. ピアソン相関係数を計算し、表示
correlation = subcategory_sales_analysis['Product_Count'].corr(subcategory_sales_analysis['Total_Sales'])
print(f"### 'Beauty & Skincare' カテゴリの製品数と総販売数のピアソン相関係数: {correlation:.2f}")
````

## 5. 価格帯分布分析による利益率向上提案

# Shopify Trending Products Profit Analysis

## 概要

このプロジェクトは、`shopify_trending_products_2025.csv` データセットを使用して、Shopifyのトレンド製品の収益性を分析し、売上最大化とコスト最適化のための戦略的洞察を特定することを目的としています。データの前処理から、カテゴリ別のパフォーマンス分析、価格帯分布の調査、そして特定のカテゴリ (`Beauty & Skincare`) の詳細分析までを行います。

## データソース

- `shopify_trending_products_2025.csv`: 2025年のShopifyトレンド製品に関するデータが含まれています。製品ID、製品名、カテゴリ、サブカテゴリ、推定販売数、推定収益、価格帯、トレンドスコアなどの情報が含まれます。

## 分析ステップ

1.  **データ読み込みと初期探索**: ZIPファイルの解凍後、CSVファイルをPandas DataFrameとして読み込み、基本的な情報（データ型、欠損値、統計量）を確認しました。
2.  **データクリーニングと前処理**: `Price_Range_USD` カラム（例: `$10-20`）を数値化し、`Min_Price_USD`, `Max_Price_USD`, `Average_Price_USD` の各カラムを作成しました。
3.  **利益関連指標の作成**: コストデータがないため、仮の粗利益率 `30%` を設定し、`Estimated_Profit_in_2025_USD` を計算しました。
4.  **カテゴリ別・サブカテゴリ別パフォーマンス分析**: `Category` および `Subcategory` ごとに、総収益、総利益、総販売数、平均価格、製品数を集計し、パフォーマンスを可視化しました。これにより、高利益カテゴリ（例: Fitness, Pet, Digital Goods, Beauty & Skincare）と低利益カテゴリ（例: Food & Beverage）を特定しました。
5.  **価格帯分布分析**: 製品全体の`Average_Price_USD`の分布をヒストグラムで可視化し、さらにカテゴリごとの平均価格の分布を箱ひげ図で分析しました。これは、価格設定の最適化や市場機会の特定に役立ちます。
6.  **Beauty & Skincare カテゴリの詳細分析**: このカテゴリに特化し、サブカテゴリごとの製品数と総販売数の関係を散布図で可視化し、ピアソン相関係数を計算して相関の強さを評価しました。

## 主要な洞察と結果

-   **高利益カテゴリ**: Fitness, Pet, Digital Goods, Beauty & Skincare が主要な収益源であることが確認されました。
-   **低利益カテゴリ**: Food & Beverage は製品数が少なく、平均価格も非常に低いため、利益貢献度が低いことが示されました。
-   **Beauty & Skincare の相関**: `Beauty & Skincare` カテゴリ内では、製品数と総販売数の間に強い正の相関 (`0.95`) が見られました。これは、製品ラインナップの拡充が売上増加に直接繋がる可能性を示唆しています。
-   **価格分布**: 全体的な価格分布とカテゴリ別の価格帯を確認することで、特定のカテゴリで価格設定の機会がある可能性が示唆されました。
